```python
import pandas as pd
df = pd.read_csv('Online retail.csv',header=None)
df.head(10)
```

```python
trans = []
for i in range(0, df.shape[0]):  # Iterate through rows using df.shape[0] for number of rows
  trans.append([str(df.values[i,j]) for j in range(0, df.shape[1])]) # Iterate through columns using df.shape[1] for number of columns
trans

len(trans)
```

```python
from apyori import apriori
results = apriori(transactions=trans,min_support = 0.003, 
                               min_confidence = 0.2, 
                               min_lift = 3, 
                               min_length = 2, 
                               max_length = 2)



report = list(results)
print(report)
print(len(report))

```

```python
# Initialize lists for storing the metrics
s = []  # Support
b = []  # Base (antecedents)
a = []  # Add (consequents)
c = []  # Confidence
l = []  # Lift

```

```python
# Loop through the results and extract the metrics
for i in range(0,9):
    s.append(report[i][1])#support
    b.append(report[i][2][0][0])#base
    a.append(report[0][2][0][1])#add
    c.append(report[i][2][0][1])#confidence
    l.append(report[i][2][0][3])#lift
```

```python
# Convert the lists into DataFrames
S_pd = pd.DataFrame(s, columns=["Support"])
B_pd = pd.DataFrame(b, columns=["Base"])
A_pd = pd.DataFrame(a, columns=["Add"])
C_pd = pd.DataFrame(c, columns=["Confidence"])
L_pd = pd.DataFrame(l, columns=["Lift"])
```

```python
# Concatenate all DataFrames into one final DataFrame
data = pd.concat([B_pd, A_pd, S_pd, C_pd, L_pd], axis=1)
data
```

```python
#save the DataFrame to a CSV file
data.to_csv("association_rules_results.csv", index=False)
```
